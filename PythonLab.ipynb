```python
import pandas as pd
import numpy as np
from IPython.display import display

df_ord = pd.read_csv('Orders.csv')
df_ret = pd.read_csv('Returns.csv')
```

```python
df_ord.columns
```

```python
x = ['Profit', 'Sales']

df_ord[x] = df_ord[x].apply(lambda x: x.str.replace('$', ''))
df_ord[x] = df_ord[x].apply(lambda x: x.str.replace(',', ''))
df_ord[x] = df_ord[x].astype('float64')
```

```python
df_ord['Year'] = pd.DatetimeIndex(df_ord['Order.Date']).year
```

```python
df_ord = df_ord.drop('Postal.Code', axis = 1)
```

```python
df_ret.shape
```

```python
df_ret.nunique()
```

```python
df_tot = pd.merge(df_ord, df_ret, left_on = 'Order.ID', right_on = 'Order ID', how = 'left')
df_tot
```

```python
df_tot = df_tot.rename(index = str, columns = {'Region_x': 'Region'})
```

```python
df_tot = df_tot.drop(['Order ID', 'Region_y'], axis = 1)
df_tot[df_tot.Returned.isnull()] = ''
```

```python
df_tot.columns
```

```python
#3.1
df_tot[df_tot.Returned == 'Yes'].groupby('Year').agg({'Sales':'sum'})
```

```python
#3.2
x = df_tot[df_tot.Returned == 'Yes'].groupby('Customer.ID').agg({'Sales':'count'})
print('More than 1 Return: ', len(x[x.Sales > 1]))
print('More than 5 Return: ', len(x[x.Sales > 5]))
```

```python
#3.3
df_tot[df_tot.Returned == 'Yes'].groupby('Region').agg({'Sales':'count'}).sort_values(by = 'Sales', ascending = False)
```

```python
#3.4
df_tot[df_tot.Returned == 'Yes'].groupby('Sub.Category').agg({'Sales':'count'}).sort_values(by = 'Sales', ascending = False)
```
